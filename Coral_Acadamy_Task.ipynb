```python
# Topics Table Cleaning

# Grouping data to check the most common topics per teacher
teacher_topic_distribution = df.groupby('Teacher Name')['Topics'].value_counts().unstack(fill_value=0)

# Print the distribution for manual review
print(teacher_topic_distribution)

# Identify the most common topic per teacher
teacher_to_topics = {}
for teacher in teacher_topic_distribution.index:
    common_topics = teacher_topic_distribution.loc[teacher].sort_values(ascending=False)
    topics = common_topics[common_topics > 0].index.tolist()  # List of topics associated with the teacher
    teacher_to_topics[teacher] = topics

# Print the preliminary mapping
for teacher, topics in teacher_to_topics.items():
    print(f"{teacher}: {topics}")
```

```python
# Calculate the frequency of each teacher-topic pair and store the counts
topic_enrollments = df.groupby(['Teacher Name', 'Topics']).size().reset_index(name='Enrollments')

# Exclude rows where 'Topics' is missing to ensure accurate frequency calculation
topic_enrollments = topic_enrollments.dropna(subset=['Topics'])

# Create a dynamic mapping of teachers to their topics, sorted by enrollment frequency
teacher_to_topics_dynamic = (
    topic_enrollments.groupby('Teacher Name')
    .apply(lambda x: x.sort_values('Enrollments', ascending=False)['Topics'].tolist())
    .to_dict()
)

# Function to impute missing topics based on the teacher's most enrolled topic
def impute_topics(row):
    if pd.isna(row['Topics']):  # Check if the topic is missing
        teacher = row['Teacher Name']
        if teacher in teacher_to_topics_dynamic:
            topics = teacher_to_topics_dynamic[teacher]
            # Select the most enrolled topic if available
            return topics[0] if topics else "Unknown Topic"
        else:
            return "Unknown Topic"  # Assign "Unknown Topic" if teacher not in mapping
    return row['Topics']  # Retain the original topic if not missing

# Fill missing values in the 'Topics' column using the dynamic mapping
df['Topics'] = df.apply(impute_topics, axis=1)
```
